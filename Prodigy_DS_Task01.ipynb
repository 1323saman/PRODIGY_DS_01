```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart
plt.figure(figsize=(8, 5))

sns.barplot(data= data, x='lcat', y='totfishing', estimator='mean', palette='viridis', errorbar=None)

plt.xlabel('Category')
plt.ylabel('Average Total Fishing Effort')
plt.title('Average Total Fishing by Category')
plt.show()
```

```python
# Histogram

plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='popden', hue='lcat', multiple='stack', bins=30, kde=False, alpha=0.6, palette='viridis')
plt.xlabel('Population Density')
plt.ylabel('Frequency')
plt.title('Histogram of Population Density by Category')
plt.show()
```
